#### Normalized Distribution (Gaussian)

```python
from scipy import stats
plt.figure(figsize=(15, 10))

# Iterate through each column and plot on a separate subplot
for i, column in enumerate(numeric_df.columns):
    plt.subplot(4, 3, i+1)
    sns.histplot(numeric_df[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xticks(rotation=45)
    
    # Check for normal distribution
    k2, p = stats.normaltest(numeric_df[column])
    #print(k2, p )
    # null hyposthesis => The sample is from Normal dsitribution
    if (p > 0.05):
        plt.text(0.5, 0.5, "Normal", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)
    else:
        plt.text(0.5, 0.5, "Not Normal", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)
    
# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

#### Skewed Distribution (negative and positive)


```python
plt.figure(figsize=(15, 10))

for i, column in enumerate(numeric_df.columns):
    plt.subplot(4, 3, i+1)
    sns.histplot(numeric_df[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xticks(rotation=45)
    
    # Check for skewness
    skewness = stats.skew(numeric_df[column])
    if skewness < -1 or skewness > 1:
        plt.text(0.5, 0.3, f"Skewed {skewness:.2f}", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)
    else:
        plt.text(0.5, 0.3, f"Not Skewed", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)


plt.tight_layout()
plt.show()
```

### Calculate Z- Score

```python
z_scores = np.abs((numeric_df - numeric_df.mean()) / numeric_df.std())
z_scores
```

```python
# Using zscore function from Scipy library
from scipy.stats import zscore

zscore = zscore(numeric_df)
```

```python
# Identify outliers (Z-score > 3)
outliers = (z_scores > 3).any(axis=1)
print("These are rows with outliers", end='\n\n')
z_scores[outliers]
```

### Identify outliers using IQR

```python
Q1,Q3 = df[column].quantile([.25,.75])
IQR = Q3-Q1
upper_limit, lower_limit = Q3 + 1.5 * IQR, Q1 - 1.5 * IQR

for column in numeric_df.columns:
    
    if any((numeric_df[column] > upper_limit) | (numeric_df[column] < lower_limit)):
        print(f"{column} has outliers.")
    else:
        print(f"{column} has no outliers.")
```

## Remove outliers

If majority of the columns are skewed then remove outliers using **IQR Method**.

```python
for column in numeric_df.columns:
    outlier_rows = numeric_df[(numeric_df[column] < lower_limit) | (numeric_df[column] > upper_limit)]
outlier_rows
```

```python
numeric_df.drop(outlier_rows.index, axis=0, inplace=True)
```

### Normal Distribution Check for Standardization and Normalization

```python
plt.figure(figsize=(15, 10))

# Iterate through each column and plot on a separate subplot
for i, column in enumerate(numeric_df.columns):
    plt.subplot(4, 3, i+1)
    sns.histplot(numeric_df[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xticks(rotation=45)
    
    # Check for normal distribution
    k2, p = stats.normaltest(numeric_df[column])
    #print(k2, p )
    # null hyposthesis => The sample is from Normal dsitribution
    if (p > 0.05):
        plt.text(0.5, 0.5, "Normal", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)
    else:
        plt.text(0.5, 0.5, "Not Normal", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)
    
# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

We will apply standardization to the columns whose distribution are normal and normalization to those where distribution is not normal.

### Standardization and Normalization

Apply standardization and normalization on the columns

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
standardizaied_columns = [] #Columns that need to be standardized
normalized_columns = [] #Columns that need to be standardized

for column in numeric_df.columns:
    _, p = stats.normaltest(numeric_df[column])
    if p>0.05:
        standardizaied_columns.append(column)
    else:
        normalized_columns.append(column)
```

```python
print(standardizaied_columns)
print(normalized_columns)
```

## Standardization

```python
numeric_df[standardizaied_columns].head()
```

```python
scaler = StandardScaler()
for col in standardizaied_columns:
    numeric_df[col] = scaler.fit_transform(numeric_df[[col]])
```

```python
numeric_df[standardizaied_columns].head()
```

## Normalization

```python
numeric_df[normalized_columns].head()
```

```python
normalize = MinMaxScaler()

for col in normalized_columns:
    numeric_df[col] = normalize.fit_transform(numeric_df[[col]])
```
