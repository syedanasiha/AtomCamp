```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# for interactive plots
# %matplotlib notebook 
```

### Simple plot with lists

```python
x = range(-10, 11)
y = [num**2 for num in x]

plt.figure(figsize = (10,5), edgecolor = "g", linewidth = 2) #all the changes to background figure/canvas are made here
plt.plot(x,y, marker = "o", color = 'r', markerfacecolor = 'k', linestyle = "--")
plt.title("My Graph", color = "r", size = 15)
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-20,20)
plt.ylim(-10, 120)

```

### Plotting on the same figure

```python
#it is much easier to do operations on arrays instead of lists so we will use np arrays now


#plotting both plots on the same figure. plt.legend() will work only if you add labels to each plot
x = np.arange(-10,11)

plt.figure()
plt.title("first plot")
plt.plot(x, x**2, "k--", label = "x vs x sq")
plt.plot(x, -x, "ro", label = "x vs -x")
plt.legend(loc = "upper center")
```

### Plotting on the different figures

```python
#to plot on separate figures, initiate a figure instance each time

x = np.arange(-10,11)

plt.figure()
plt.plot(x, x**2, "k--", label = "x vs x sq")

plt.figure()
plt.plot(x, -x, "ro", label = "x vs -x")
```

### Storing your plot and axes in variables to plot on a figure from a different cell

```python
#you can store your figure and axes in variables. this way, you can call on them/ plot them from a separate cell too. 

x = np.arange(-10,11)

fig = plt.figure()
axes = plt.axes()


axes.plot(x, -x, "r")
```

```python
#plotting from a separate cell on the plot above

axes.plot(x, x**2, "b")
fig
```

## plt.subplots() 
can be used to generate one plot or multiple subplots on the same figure <br>
this method returns both figure and axes to you, so you don't need to initiate them separately like above <br>
often used to plot single graphs because it returns both figure and axes

##### NOTE:
with plt.plot() your commands are plt.title(), plt.xlabel(), plt.ylim() etc. <br>
when you store your axes in a variable, and call on axes, then the commands become ax.set_title(), ax.set_xlabel(), ax.set_ylim() etc.


```python
x = np.arange(-10,11)
fig, axes = plt.subplots(figsize= (8,5), linewidth = 2, facecolor = "g") #all the figure changes are made here

axes.plot(x, -x, "r") 
axes.set_title("x vs -x", color = 'w', loc = 'center', size = 15)
axes.set_xlabel("-x", color = 'w', size = 15)
axes.set_ylabel("x", color = 'w', size = 15)
```

```python
#you can create more than one subplots in the same figure

fig, axes = plt.subplots(2,3, figsize= (10,8), edgecolor = "b", linewidth = 2)



axes[0,0].plot(x,-x, "r")
axes[0,0].set_title("subplot 1")
axes[0,0].set_xlabel("-x", size = 15)
axes[0,0].set_ylabel("x", size = 15)

axes[1,1].plot(x, x**2, "g")
axes[1,1].set_title("subplot 5")


#try plotting something on the following axes
# axes[0,1]
# axes[1,2]

```

```python
#you can also unpack the subplots like this and assign each to its own variable 

x = np.arange(-10,11)
fig, ((ax1, ax2, ax3), (ax4,ax5,ax6)) = plt.subplots(2,3, figsize= (10,8), edgecolor = "b", linewidth = 2)


ax2.plot(x,-x, "r")
ax2.set_title("y = -x", color = "r", size = 10)
ax2.set_xlabel("-x")
ax2.set_ylabel("y")
ax2.yaxis.set_label_coords(-0.1, 0.5) #this is to move the position of y_label. i will show you how to get these coordinates. 
ax2.set_xlim(-12, 12)
ax2.set_ylim (-12, 12)

ax6.plot(x, x**2, "g")
ax6.set_title("y = x**2", color = "b", size = 10)
ax6.set_xlabel("x")
ax6.set_ylabel("y") #y_label overlaps with ax 5 in the plot. lets get coordinates of this below
ax6.set_xlim(-12, 12)
ax6.set_ylim (0, 110)







#as you can see from the code above, the y_label of ax6 overlaps with the ax5 plot
#For EACH subplot, the default coordinates of x_label and y_label are (0, 0.5) where (0,0) is the bottom left of the subplot and (0.5, 0.5) is the center
#lets see how we can see the coordinates of our y_label. 

#to get the position coordinates: 
#you can either store the set_ylabel command in a variable and call on that variable to get the position as follows:
ylbl = ax6.set_ylabel("y")
ylbl.get_position()

#or you can directly do it like below
ax6.set_ylabel("y").get_position()


#simply adjust these coordinates with axes.yaxis.set_label_coords() or axes.xaxis.set_label_coords()
# ax6.yaxis.set_label_coords(-0.1, 0.5)
```

### SCATTER PLOT

```python
x = np.array([1,2,3,4])
y = np.array([1,2,3,4])

sizes = np.array([10,20,30,40])
colors = np.random.rand(4)

plt.scatter(x,y, s = sizes*10, c = colors, alpha = 0.9, cmap = "Spectral")


#plt.scatter is taking in the following
# 1) a list of 4 x values
# 2) a list of 4 y values
# 3) a list of 4 numbers for s which determines size of each point
# 4) a list of 4 numbers for colors (that we randomly generated and stored in the variable colors)
# 5) a single value for alpha, which is the transparency
# 6) a single value for cmap. cmap is the color theme matplotlib will use to map your colors array (which contains numbers here) to colors

# it will now cycle through the lists, grabbing x[0], y[0] as the point to plot, will make area[0] its size and color[0] its color
# then it'll do the same for the next set of points i.e. x[1],y[1],area[1],color[1] and so on
```

```python
income = np.array([50000, 50000, 55000, 70000, 80000, 75000, 65000, 90000, 85000, 82000, 100000, 110000, 105000, 120000, 115000])
children = np.array([2, 3, 1, 4, 2, 3, 1, 5, 3, 2, 4, 3, 2, 5, 4])
house_size = np.array([100, 200, 100, 1000, 2200, 2100, 1900, 2500, 2400, 5000, 2600, 8000, 2200, 10000, 5500])

colors = np.arange(0,15)

plt.scatter(income, children, s = house_size, c = colors, alpha = 0.5, edgecolors = "k", cmap = "Spectral")
plt.title("Housesize")

```

#### SETTING A LEGEND FOR A SCATTER PLOT

```python
fig, ax1 = plt.subplots(figsize= (5, 5)) #creating a single plot with plt.subplots first


x= np.random.randint(0,80,1500) #1500 random integers between 0 and 80 (not including 80)
y= np.random.randint(0,250,1500) #1500 random integers between 0 and 250 (not including 250)
colors = np.random.randint(0,3,len(x)) #1500 random integers between 0 and 3 (not including 3) so 0, 1 or 2

#now we will plot on ax1 but we will store the plot in a variable
scatter = ax1.scatter(x, y, c = colors)
ax1.set_title("random")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_xlim(0, 100)
ax1.set_ylim(0, 300)


#now we use the variable scatter that is storing our plot to get legend_elements, which returns a tuple that we will unpack with x, y
x, y = scatter.legend_elements()


#if you print x and y, you will observe that x stores the 2-D objects and y stores a list
print(x)
print(y)

#pass x and y to ax1.legend to get the label of each color
ax1.legend(x, y)

#if you want the labels to be something else for e.g if 0 is "alpha", 1 is "beta" and 2 is "gamma", you can specify them in the order of the list y
#y is ['$\\mathdefault{0}$', '$\\mathdefault{1}$', '$\\mathdefault{2}$'] 
#pass x as the first argument and your labels as the second argument in the same order as contained in y like below:

# ax1.legend(x, ["alpha", "beta", "gamma"])
```

## subplot2grid

```python
ax1 = plt.subplot2grid((3,3), (0,0), rowspan = 2, colspan = 3)
ax2 = plt.subplot2grid((3,3), (2,0))
ax3 = plt.subplot2grid((3,3), (2,1), colspan = 2 )


#first argument (3,3) is a tuple specifying the arrangement of subplots. this means 3 rows and 3 columns
#second argment (0,0) is a tuple of position coordinates for the subplot within the larger arrangement
# rowspan and colspan tells how many rows and columns you want the subplot to span
```

#### STACKED BAR

```python
#bargraph of random values

Y = np.random.rand(5) #5 randomly generated values

plt.figure(figsize=(8, 2))

plt.bar(np.arange(len(Y)), Y, width=0.3, color='#00b894') 

#x values are passed as np.arange(len(Y)). this function will generate integers starting from 0, uptil (but not including) the length of Y. 
#here, the length of Y is 5, so np.arange(5) is giving you 0,1,2,3,4 as you can see on the x-axis

```

```python
#this is a stacked bar. the code is the same as above but we are adding a second plot command for Y2
#in the plot command for Y2, you will need to specify the bottom argument and set it to Y, 
#then Y2 is plotted at the top and Y is the bottom

Y = np.random.rand(5) #5 randomly generated values
Y2 = np.random.rand(5) #again 5 randomly generated values

plt.figure(figsize=(8, 4))

barWidth = 0.5
plt.bar(np.arange(len(Y)), Y, width=barWidth, color='#00b894', label='Label Y')
plt.bar(np.arange(len(Y2)), Y2, width=barWidth, color='#e17055', bottom=Y, label='Label Y2')

plt.legend()
plt.show()
```
